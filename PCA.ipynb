```python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mnist_images, mnist_labels = fetch_openml("mnist_784", version= 1, return_X_y=True, as_frame=False)
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(mnist_images[-1].reshape(28,28))
plt.title(f"le chiffre {mnist_labels[-1]}")
plt.show()
```

```python
plt.figure(figsize=(16, 6))
sns.histplot(mnist_labels, discrete=True)
plt.title("MNIST Label Distribution")

plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(mnist_images)
mnist_images_scalled = scaler.transform(mnist_images)
```

Premier utlisation du la PCA

```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
pca_object =PCA()
pca_object.fit(mnist_images_scalled)
```

```python
explained_variances = pca_object.explained_variance_ratio_#C'est liste des lamba i divise par la trace(poids relatifs ou proportions de poids)
trace = 28*28
cumulatif_explained_variance = np.cumsum(explained_variances)
nb_composant_to_save = np.argmax(cumulatif_explained_variance > 0.95)+1
```

```python
explained_variances * trace
```

```python
plt.figure(figsize=(16, 6))
plt.plot(cumulatif_explained_variance)
plt.scatter(np.argmax(cumulatif_explained_variance> 0.95), 0.95, label= "Seuil pour garder 95 % de la qualite et le poids informationnelle")
plt.show()
```
